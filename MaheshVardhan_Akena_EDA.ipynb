```python
# On customers data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Data/Customers.csv")

print("Dataset Preview:")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe(include='all'))
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Rows:")
print(df.duplicated().sum())



df = df.drop_duplicates()
df['SignupDate'] = pd.to_datetime(df['SignupDate'])
df['SignupYear'] = df['SignupDate'].dt.year
df['SignupMonth'] = df['SignupDate'].dt.month
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='SignupYear', palette='viridis')
plt.title('Number of Customers by Signup Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='SignupMonth', palette='plasma')
plt.title('Signup Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()

if 'Region' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='Region', palette='cool')
    plt.title('Customers by Region')
    plt.xlabel('Region')
    plt.ylabel('Count')
    plt.show()

df_grouped = df.groupby('SignupYear').size().reset_index(name='CustomerCount')
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_grouped, x='SignupYear', y='CustomerCount', marker='o')
plt.title('Customer Signup Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.show()

if 'Sales' in df.columns and 'Profit' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='Sales', y='Profit', hue='Segment', palette='cool', alpha=0.7)
    plt.title('Sales vs. Profit Analysis')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.show()

print("EDA Completed.")
```

```python
# EDA on product data
df = pd.read_csv("/content/Data/Products.csv")

print(df.head())
print(df.info())
print(df.describe(include='all'))
print(df.isnull().sum())
print(df.duplicated().sum())
df = df.drop_duplicates()
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Category', palette='viridis', order=df['Category'].value_counts().index)
plt.title('Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(8, 6))
sns.histplot(df['Price'], bins=20, kde=True, color='blue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()

avg_price_per_category = df.groupby('Category')['Price'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_price_per_category, x='Category', y='Price', palette='plasma', order=avg_price_per_category.sort_values('Price', ascending=False)['Category'])
plt.title('Average Price by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()



plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Price', palette='cool')
plt.title('Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.show()
top_10_expensive = df.nlargest(10, 'Price')
plt.figure(figsize=(8, 6))
sns.barplot(data=top_10_expensive, x='ProductName', y='Price', palette='magma')
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Product Name')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.show()

```

```python

df = pd.read_csv("Transactions.csv")

print("Dataset Preview:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe(include='all'))

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

df = df.drop_duplicates()
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['Year'] = df['TransactionDate'].dt.year
df['Month'] = df['TransactionDate'].dt.month
df['Day'] = df['TransactionDate'].dt.day
df['Weekday'] = df['TransactionDate'].dt.weekday
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Number of Transactions by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Month', palette='plasma')
plt.title('Transactions by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
monthly_sales = df.groupby(['Year', 'Month'])['TotalValue'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='TotalValue', hue='Year', marker='o')
plt.title('Total Transaction Value by Month')
plt.xlabel('Month')
plt.ylabel('Total Value (USD)')
plt.show()
quantity_by_product = df.groupby('ProductID')['Quantity'].sum().reset_index()
quantity_by_product = quantity_by_product.sort_values('Quantity', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=quantity_by_product.head(10), x='ProductID', y='Quantity', palette='magma')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product ID')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

total_value_by_product = df.groupby('ProductID')['TotalValue'].sum().reset_index()
total_value_by_product = total_value_by_product.sort_values('TotalValue', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=total_value_by_product.head(10), x='ProductID', y='TotalValue', palette='coolwarm')
plt.title('Top 10 Products by Total Sales Value')
plt.xlabel('Product ID')
plt.ylabel('Total Sales Value (USD)')
plt.xticks(rotation=45)
plt.show()
```
